## String basics

Strings are immutable. Use single or double quotes — or triple-quoted for multi-line.

````python
msg = "slack violation"
multi = '''line1
line2'''
print(msg, len(msg))
print(multi)
````

## f-strings with format specs

`{value:spec}` gives you padding, precision, sign, thousands separators.

````python
name, slack, cap = "reg2reg", -0.045, 1.25e-12
print(f"{name:10s} slack={slack:+.3f} ns cap={cap:.2e} F")
print(f"hex addr: {255:#06x}  float: {3.14:>10.4f}")
````

## Indexing and slicing

Strings support negative indices and `start:stop:step` slicing.

````python
path = "top/cpu/alu/adder_0"
print(path[0], path[-1])
print(path[:3], path[-7:])
print(path[::-1])
````

## Splitting and joining

`.split(sep)` breaks; `'/'.join(parts)` stitches back together.

````python
hier = "top/cpu/alu/adder_0"
parts = hier.split("/")
print("parts  :", parts)
print("parent :", "/".join(parts[:-1]))
print("leaf   :", parts[-1])
````

## Searching and replacing

`in`, `startswith`, `endswith`, `find`, `replace` cover most text tasks.

````python
line = "slack -0.045 ns setup violation"
print("has violation :", "violation" in line)
print("index of ns   :", line.find("ns"))
print(line.replace("violation", "VIOL"))
````

## Cleaning whitespace

`strip`, `lstrip`, `rstrip`, `splitlines` make raw logs presentable.

````python
raw = "   /path/to/top/U1   \n"
print(repr(raw.strip()))

report = "line1\n  line2  \n\nline3\n"
for l in report.splitlines():
    clean = l.strip()
    if clean:
        print(repr(clean))
````

## Regex essentials

`re.search` / `re.findall` give you pattern-based extraction without brittle `split` chains.

````python
import re
line = "instance U0/cpu/alu/and2_ (path=top/cpu/alu/and2_) slack=-0.045ns"

m = re.search(r"slack=(-?\d+\.\d+)ns", line)
print("slack :", float(m.group(1)) if m else None)

paths = re.findall(r"top/\S+", line)
print("paths :", paths)
````

## VLSI practice — hierarchical path utilities

Turn a Cadence-style `top/U1/U2/inst` string into leaf, parent, and depth — the kind of helpers every flow script needs.

````python
def leaf(path):   return path.rsplit("/", 1)[-1]
def parent(path):
    h, _, _ = path.rpartition("/")
    return h or None
def depth(path):  return path.count("/") + 1

for p in ("top", "top/cpu/alu/and2_", "top/cpu"):
    print(f"{p:22s} leaf={leaf(p):10s} parent={parent(p)}  depth={depth(p)}")
````

## Recap

- Strings are immutable; every method returns a new string.
- f-strings are the cleanest way to format reports.
- `split`/`join` and `in`/`startswith` handle most text slicing.
- Reach for `re` when patterns vary — don't force `split` to be regex.
