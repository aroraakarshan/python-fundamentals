## The same task, three languages

The book opens with the classic example: open a report file and print every line. Watch how much noise Python strips away compared to Tcl and Perl.

**Tcl** (reference only):

```tcl
set file [open "report.txt" r]
while {[gets $file line] != -1} {
    puts $line
}
close $file
```

**Perl** (reference only):

```perl
open(my $fh, '<', "report.txt") or die "Cannot open file: $!";
while (my $line = <$fh>) {
    chomp $line;
    print "$line\n";
}
close($fh);
```

**Python** — and this one actually runs:

````python
from pathlib import Path

# Create a tiny sample report so the cell is self-contained.
Path('report.txt').write_text(
    'Startpoint: U1/clk\n'
    'Endpoint: U2/data\n'
    'Slack: -0.045 setup\n'
)

with open('report.txt') as file:
    for line in file:
        print(line.strip())
````

Three lines of real work. No `set`, no `$`, no manual `close`. The `with` block closes the file for you when the loop ends.

## Why this matters beyond readability

The book calls out four things Python gives you that the older languages don't:

1. **A massive standard library** — `os`, `subprocess`, `re`, `pathlib`, `json` ship with every install.
2. **A huge ecosystem of free packages** — `pandas`, `numpy`, `matplotlib`, `click`, `jinja2`, `streamlit`.
3. **First-class support in modern EDA tools** — Synopsys, Cadence, and ANSYS expose Python APIs.
4. **AI assistants produce better Python than Tcl or Perl.** In practice, you move faster.

````python
# Automating a shell command is one line in Python.
import subprocess, sys
result = subprocess.run([sys.executable, '--version'], capture_output=True, text=True)
print('stdout:', result.stdout.strip() or result.stderr.strip())
print('return code:', result.returncode)
````

The book uses `os.system("calibre -drc my_design.gds")` as the illustration — `subprocess.run` is the modern, safer replacement and captures output for you.
