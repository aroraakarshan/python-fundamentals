## Python in the VLSI flow

Every major EDA vendor (Synopsys, Cadence, Siemens) ships Python bindings. Engineers use it to glue flows, parse reports, mine constraints, and drive regressions. Compared to Tcl its syntax is friendlier; compared to Perl its data structures are cleaner.

````python
languages = ["Tcl", "Perl", "Python"]
strengths = {
    "Tcl":    "Native to most EDA tools",
    "Perl":   "Legacy text mangling",
    "Python": "Readable, huge stdlib, data science ecosystem",
}
for lang in languages:
    print(f"{lang:7s} -> {strengths[lang]}")
````

## Your first interactive expressions

Python runs top to bottom. `print` is how you see values; plain expressions at the end of a cell also echo in Jupyter, but `print` is always safe.

````python
design_name = "cpu_top"
num_cells   = 125_430
freq_ghz    = 1.25

print("design :", design_name)
print("cells  :", num_cells)
print("period :", round(1 / freq_ghz, 3), "ns")
````

## Readable arithmetic

Python supports `+ - * /` plus `//` for integer division, `%` for modulus, and `**` for exponent. Underscores in numbers are allowed for readability.

````python
gates     = 1_250_000
die_mm2   = 12.5
density   = gates / die_mm2
print(f"density = {density:,.0f} gates / mm^2")
print("2**10   =", 2 ** 10)
print("17 % 5  =", 17 % 5)
print("17 // 5 =", 17 // 5)
````

## Strings are first-class

VLSI work is mostly text: netlists, SDCs, reports. f-strings embed values cleanly.

````python
cell_name = "INV_X1"
pins = ["A", "Y", "VDD", "VSS"]
print(f"{cell_name} has {len(pins)} pins: {pins}")
print(cell_name.lower(), cell_name.replace('_X1', '_X2'))
````

## Lists and dicts in one breath

Lists are ordered sequences. Dicts map keys to values. Both feel natural once you run them.

````python
critical_nets = ["clk", "reset", "scan_en"]
fanouts = {"clk": 4200, "reset": 800, "scan_en": 1200}

for net in critical_nets:
    print(f"{net:8s} fanout = {fanouts[net]:,}")
````

## Why this matters in practice

A ten-line Python loop often replaces a hundred-line Tcl procedure. Legible code means fewer bugs on the critical path to tapeout.

````python
violations = [("reg_a2b", -0.045), ("io_out", -0.002), ("reg_c2d", -0.110)]
violations.sort(key=lambda r: r[1])
for name, slack in violations[:2]:
    print(f"worst: {name:10s} slack={slack:+.3f} ns")
````

## VLSI practice — first report summary

Given a tiny list of slack values, count violations and report the worst.

````python
slacks = [0.12, -0.045, 0.004, -0.110, 0.080, -0.002]
violations = [s for s in slacks if s < 0]
print(f"total paths : {len(slacks)}")
print(f"violations  : {len(violations)}")
print(f"worst slack : {min(slacks):+.3f} ns")
````

## Recap

- Python is readable, ubiquitous in EDA, and great for flow glue.
- `print` and f-strings handle almost all reporting needs.
- Lists and dicts replace most ad-hoc data structures.
- A short Python loop often replaces many lines of Tcl/Perl.
