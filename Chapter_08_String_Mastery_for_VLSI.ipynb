## Regular expressions for report parsing

````python
import re
line = 'Slack: -0.034 ns'
match = re.search(r'Slack:\s*(-?\d+\.\d+)\s*ns', line)
if match:
    slack = float(match.group(1))
    print('slack =', slack)
````

## VLSI in practice — summarise a short STA report

````python
from pathlib import Path
Path('sta.rpt').write_text(
    'Startpoint: U1/clk\n'
    'Endpoint: U2/data\n'
    'slack -0.045 setup violation\n'
    'slack 0.120 hold met\n'
    'slack -0.089 setup violation\n'
)
violations = []
with open('sta.rpt') as f:
    for line in f:
        if line.startswith('slack'):
            tokens = line.split()
            slack = float(tokens[1])
            if slack < 0:
                violations.append(slack)
violations.sort()
print('worst slack:', violations[0] if violations else 'none')
print(f'{len(violations)} violation(s)')
````

## Recap

- Strings are immutable sequences; operations return new strings.
- Prefer `f-strings` — the clearest formatting in modern Python.
- `split`, `join`, `replace`, `startswith` cover 80% of report parsing.
- Reach for `re.search` when the pattern isn't a simple delimiter.
