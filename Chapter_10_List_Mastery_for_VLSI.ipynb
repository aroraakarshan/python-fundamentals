## Creating lists

````python
cells = ['INV_X1', 'NAND2_X2', 'DFF_X1']
slack_values = [0.1, -0.05, 0.0]
print(cells, slack_values)
````

## Access and modify

````python
cells = ['U1', 'U2', 'U3']
print(cells[0], cells[-1])
cells[1] = 'U2_MOD'
print(cells)
````

## Add and remove

````python
cells = ['U1', 'U2', 'U3']
cells.append('U4')
cells.extend(['U5', 'U6'])
if 'U2' in cells:
    cells.remove('U2')
last = cells.pop()
cells.insert(1, 'U_NEW')
print(cells, 'popped:', last)
````

## Looping, length, membership

````python
cells = ['U1', 'U2', 'U3']
for c in cells:
    print('checking', c)
print('count:', len(cells))
print('U1 in cells?', 'U1' in cells)
````

## Sorting — beware of lexicographic order

`U10` sorts before `U2` by default. Pass a `key` to sort by the numeric part.

````python
cells = ['U10', 'U2', 'U1']
cells.sort()
print('lexical:', cells)
cells.sort(key=lambda x: int(x[1:]))
print('numeric:', cells)
slacks = [0.2, -0.1, 0.05]
slacks.sort()
print('slacks:', slacks)
````

## List comprehensions

````python
cells = ['INV_X1', 'DFF_X1', 'INV_X2', 'NAND2_X1']
inv_cells = [c for c in cells if c.startswith('INV')]
print(inv_cells)
upper = [c.lower() for c in cells]
print(upper)
````

## VLSI in practice — bucket cells by family

````python
netlist = ['INV_X1', 'NAND2_X1', 'DFF_X2', 'INV_X2', 'NOR2_X1', 'DFF_X1']
families = {}
for cell in netlist:
    prefix = cell.split('_')[0]
    families.setdefault(prefix, []).append(cell)
for family, members in families.items():
    print(f'{family}: {members}')
````

## Recap

- Lists are ordered, mutable, 0-indexed.
- `append` adds one; `extend` adds many.
- Default sort is lexicographic — pass `key=` for numeric-aware sorts.
- List comprehensions build new lists in one readable expression.
