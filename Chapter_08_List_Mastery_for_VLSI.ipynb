## Creating and indexing

Lists are ordered, mutable, and allow duplicates.

````python
cells = ["INV_X1", "NAND2_X1", "DFF_X1", "INV_X1"]
print(cells[0], cells[-1], len(cells), cells.count("INV_X1"))
````

## Slicing

`start:stop:step` gives you sub-lists without loops.

````python
a = list(range(10))
print(a[2:7], a[::2], a[::-1])
````

## Mutating methods

`append`, `extend`, `insert`, `remove`, `pop`, `sort`, `reverse`. These modify the list in place and return `None`.

````python
nets = ["clk"]
nets.append("reset")
nets.extend(["data", "scan_en"])
nets.insert(1, "pll_lock")
nets.remove("data")
print(nets)
popped = nets.pop()
print("popped:", popped, "remaining:", nets)
````

## List comprehensions

Map + filter in one expression. Prefer them over manual append loops.

````python
cells = ["INV_X1", "NAND2_X1", "DFF_X1", "MUX2_X1"]
flops = [c for c in cells if c.startswith("DFF")]
lower = [c.lower() for c in cells]
print(flops, lower)
````

## Nested lists and enumerate

Grids and tables map naturally to lists of lists.

````python
grid = [[0]*3 for _ in range(3)]
grid[1][2] = 1
for r, row in enumerate(grid):
    print(r, row)
````

## Sorting with a key

`sorted(seq, key=...)` returns a new list; `.sort()` mutates in place.

````python
paths = [("reg_a", -0.045), ("io_b", 0.12), ("reg_c", -0.110), ("mux_d", -0.002)]
by_slack = sorted(paths, key=lambda r: r[1])
print(by_slack)
print(sorted(paths, key=lambda r: abs(r[1]), reverse=True))
````

## Aliasing vs copying

Assignment does not copy. Use `list.copy()` or `[:]` to avoid surprises.

````python
a = [1, 2, 3]
b = a
b.append(99)
print("shared :", a)

c = a.copy()
c.append(-1)
print("after copy:", a, c)
````

## VLSI practice — fanout histogram

Group nets into fanout buckets and print a small histogram.

````python
nets = [("clk", 4200), ("reset", 800), ("data_0", 150), ("data_1", 180),
        ("scan", 1000), ("pll_lock", 50), ("bank_sel", 400)]

def bucket(fo):
    if fo < 100:    return "<100"
    if fo < 500:    return "100-499"
    if fo < 1000:   return "500-999"
    return ">=1000"

from collections import defaultdict
buckets = defaultdict(list)
for name, fo in nets:
    buckets[bucket(fo)].append(name)

for b in ("<100", "100-499", "500-999", ">=1000"):
    names = buckets.get(b, [])
    print(f"{b:8s} ({len(names):2d}) {names}")
````
