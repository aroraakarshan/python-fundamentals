## Creating a dictionary

````python
cell_area = {
    'INV_X1': 1.2,
    'NAND2_X1': 2.3,
    'DFF_X1': 5.1,
}
print(cell_area)
````

## Access, safely

`d[key]` raises `KeyError` if missing. `d.get(key, default)` returns the default.

````python
print(cell_area['INV_X1'])
print(cell_area.get('XYZ', 0.0))
````

## Add, modify, delete

````python
cell_area['BUF_X1'] = 0.9
cell_area['INV_X1'] = 1.1
del cell_area['NAND2_X1']
removed = cell_area.pop('DFF_X1')
print(cell_area, 'removed:', removed)
````

## `.keys()`, `.values()`, `.items()`

````python
cell_power = {'INV_X1': 0.05, 'NAND2_X1': 0.08, 'BUF_X1': 0.04}
print(list(cell_power.keys()))
print(list(cell_power.values()))
for cell, power in cell_power.items():
    print(f'{cell} uses {power} mW')
````

## `.update()` for merging

````python
cell_power.update({'DFF_X1': 0.12, 'BUF_X1': 0.045})
print(cell_power)
````

## Nested dictionaries

````python
cell_info = {
    'INV_X1': {'area': 1.2, 'power': 0.05, 'pins': 2},
    'DFF_X1': {'area': 5.1, 'power': 0.12, 'pins': 4},
}
print(cell_info['INV_X1']['area'])
for cell, info in cell_info.items():
    print(f"{cell}: area={info['area']} pins={info['pins']}")
````

## VLSI in practice — count violations by type

````python
violations = ['setup', 'hold', 'setup', 'drc', 'setup', 'hold', 'antenna']
counts = {}
for v in violations:
    counts[v] = counts.get(v, 0) + 1
for name, n in sorted(counts.items()):
    print(f'{name:8s} {n}')
````

pairs = [('INV_X1', 0.05), ('DFF_X1', 0.12), ('INV_X1', 0.07)]

    maxp[cell] = max(maxp.get(cell, 0.0), p)

## Recap

- Dictionaries map unique keys to values; access is `O(1)`.
- `.get(key, default)` avoids `KeyError`.
- `.keys()`, `.values()`, `.items()` give iterable views.
- Nest dicts for structured records; merge with `.update()`.
