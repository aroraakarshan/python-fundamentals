## Creating sets

Use `{ ... }` with at least one element, or `set(...)`. An empty `{}` is a dict.

````python
layers = {'M1', 'M2', 'M3'}
print(layers)
dup_list = ['M1', 'M2', 'M1', 'M2']
print(set(dup_list))
empty = set()
also_empty = {}
print(type(empty).__name__, type(also_empty).__name__)
````

## Add, remove, membership

````python
layers = {'M1', 'M2', 'M3'}
layers.add('M4')
layers.discard('M5')
layers.remove('M2')
print('M1' in layers)
print(layers)
````

## Union, intersection, difference

````python
layout_layers = {'M1', 'M2', 'M3'}
allowed_layers = {'M1', 'M2', 'M4'}
print('union:       ', layout_layers | allowed_layers)
print('intersection:', layout_layers & allowed_layers)
print('diff (used-allowed):', layout_layers - allowed_layers)
print('diff (allowed-used):', allowed_layers - layout_layers)
````

## VLSI in practice — find unused and illegal layers

````python
spec_layers = {'M1', 'M2', 'M3', 'M4', 'VIA12', 'VIA23'}
seen_layers = {'M1', 'M2', 'VIA12', 'POLY'}
unused = spec_layers - seen_layers
illegal = seen_layers - spec_layers
print('unused in design:', unused)
print('illegal layers:  ', illegal)
````

## Why a set is fast

Membership in a `set` is `O(1)` average; in a `list` it's `O(n)`. The difference is orders of magnitude.

````python
import time
big_list = list(range(100_000))
big_set = set(big_list)
target = 99_999
t0 = time.perf_counter(); _ = target in big_list; t_list = time.perf_counter() - t0
t0 = time.perf_counter(); _ = target in big_set;  t_set  = time.perf_counter() - t0
print(f'list lookup: {t_list*1e6:.1f} us')
print(f'set  lookup: {t_set*1e6:.3f} us')
````

## Recap

- A set is an unordered, duplicate-free collection.
- Empty set is `set()`; `{}` is an empty dict.
- Union `|`, intersection `&`, difference `-` map directly to design questions.
- Use sets for fast membership tests on large collections.
