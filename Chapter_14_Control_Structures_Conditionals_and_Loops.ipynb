## Conditionals: `if`, `elif`, `else`

The book's canonical example is classifying a timing slack.

````python
slack = -0.05
if slack < 0:
    print('Timing violation')
elif slack == 0:
    print('Critical path')
else:
    print('Path is safe')
````

Common beginner mistakes the book calls out:

- Forgetting the `:` at the end of `if`.
- Mixing tabs and spaces — Python uses indentation to group code.
- Writing `=` (assignment) when you meant `==` (comparison).

## Comparison operators

`==`, `!=`, `>`, `<`, `>=`, `<=` — each returns a `bool`.

````python
violation_count = 5
print('Under limit:', violation_count <= 10)
wire_length = 120
print('Short wire:', wire_length < 100)
````

## Logical operators: `and`, `or`, `not`

`and` has higher precedence than `or`. Use parentheses when you aren't sure.

````python
a, b, c = 1, 3, 10
print((a > 0 and b < 5) or c == 10)
print(not (a > 0))
````

## `for` loops

````python
cells = ['U1', 'U2', 'U3']
for cell in cells:
    print('Checking cell:', cell)
````

**Warning from the book:** don't modify a list while iterating over it. Iterate over a copy (`cells[:]`) or build a new list.

## `range()` — stop value is exclusive

````python
for i in range(3):
    print(i)
print('---')
for i in range(2, 8, 2):
    print('layer M', i)
````

## `while` loops

Run while the condition stays true. Make sure the condition will eventually become false!

````python
count = 0
while count < 5:
    print(count)
    count += 1
````

## `break`, `continue`, `pass`

````python
nets = ['clk', 'vdd', 'data', 'VDD', 'reset']
for net in nets:
    if net.lower() == 'vdd':
        print('stop at power net'); break
    if net.startswith('clk'):
        continue
    print('processing', net)
````

## `match` (Python 3.10+)

Cleaner than a stack of `elif`s when dispatching on a single value.

````python
def describe(layer):
    match layer:
        case 'M1': return 'Metal1 — local routing'
        case 'M2': return 'Metal2 — short interconnect'
        case 'M3': return 'Metal3 — intermediate'
        case _:    return 'Unknown layer'

for layer in ('M1', 'M2', 'M7'):
    print(layer, '→', describe(layer))
````

## VLSI in practice — flag setup violations

````python
report = [
    ('setup', -0.045),
    ('hold',   0.120),
    ('setup', -0.089),
    ('setup',  0.010),
    ('hold',  -0.005),
]
setup_v, hold_v = 0, 0
for check, slack in report:
    if slack >= 0:
        continue
    if check == 'setup':
        setup_v += 1
    else:
        hold_v += 1
print(f'setup violations: {setup_v}')
print(f'hold  violations: {hold_v}')
````
