## if / elif / else

Classify a single slack value through a chain of branches.

````python
def classify(slack):
    if slack < 0:
        return "violation"
    elif slack == 0:
        return "critical"
    elif slack < 0.05:
        return "tight"
    else:
        return "safe"

for s in (-0.1, 0.0, 0.02, 0.25):
    print(f"{s:+.3f} -> {classify(s)}")
````

## Boolean combinations

`and`, `or`, `not`. Use parentheses to keep intent obvious.

````python
slack, drc_clean, power_ok = -0.02, True, True
ship_block = (slack >= 0) and drc_clean and power_ok
print("ready to ship?", ship_block)
````

## for loops over sequences

Iterate over lists, tuples, dict keys/values/items.

````python
nets = ["clk", "reset", "data", "scan_en"]
for idx, net in enumerate(nets, start=1):
    print(f"{idx:2d}. {net}")

fanouts = {"clk": 4200, "reset": 800, "data": 150}
for net, fo in fanouts.items():
    print(f"{net:6s} -> {fo:,}")
````

## range and counting

`range(start, stop, step)` is the Pythonic way to count.

````python
for bank in range(0, 8, 2):
    print(f"bank{bank}")
````

## while loops

Use when you don't know the iteration count up front — for example, retrying a convergence step.

````python
slack    = -0.2
attempts = 0
while slack < 0 and attempts < 5:
    slack += 0.05
    attempts += 1
    print(f"attempt {attempts}: slack={slack:+.3f}")
print("done in", attempts, "attempts")
````

## break, continue, and for-else

`break` exits early, `continue` skips to next iter, the rare `for-else` runs only when the loop ended *without* a `break`.

````python
paths = [-0.1, -0.05, 0.02, -0.2]

# Find first violator worse than -0.1:
for s in paths:
    if s > -0.1:
        continue
    print("first severe violator:", s)
    break
else:
    print("no severe violator found")
````

## List comprehensions

A loop + transform + filter in one readable line.

````python
slacks = [0.12, -0.045, 0.004, -0.110, 0.080, -0.002]
violations = [s for s in slacks if s < 0]
abs_slack  = [abs(s) for s in slacks]
print("viols :", violations)
print("|s|   :", abs_slack)
````

## Dict comprehensions

Build a lookup in one line.

````python
nets = ["clk", "reset", "data"]
load = {n: round(0.1 + 0.2 * i, 2) for i, n in enumerate(nets)}
print(load)
````

## VLSI practice — worst-N slack extractor

Loop through a list of (path, slack) tuples, keep only violators, and print the worst three sorted by slack ascending.

````python
paths = [("reg_a2b", -0.045), ("io_out", -0.002), ("reg_c2d", -0.110),
         ("in2reg", 0.120), ("mux_pd", -0.080)]

violators = [p for p in paths if p[1] < 0]
violators.sort(key=lambda r: r[1])

print(f"{len(violators)} violations")
for i, (name, s) in enumerate(violators[:3], 1):
    print(f"{i}. {name:10s} slack = {s:+.3f} ns")
````
