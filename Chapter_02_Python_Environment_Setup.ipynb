## Which Python is this?

Always confirm the interpreter first — EDA servers often ship multiple.

````python
import sys, platform
print("executable :", sys.executable)
print("version    :", sys.version.replace("\n", " "))
print("platform   :", platform.platform())
print("implementation:", platform.python_implementation())
````

## Runtime paths

`sys.path` tells you where Python looks for modules. When an import fails mysteriously, this is the first thing to inspect.

````python
import sys
for i, p in enumerate(sys.path[:6]):
    print(f"{i}. {p or '(cwd)'}")
print("... and", max(0, len(sys.path) - 6), "more")
````

## Virtual environments (conceptual)

On a shared EDA server you don't want to break the system Python. A venv creates a private sandbox:

```bash
python3 -m venv .venv
source .venv/bin/activate
pip install --upgrade pip
```

We simulate the check here without actually creating a venv.

````python
import os, sys
in_venv = sys.prefix != getattr(sys, "base_prefix", sys.prefix)
print("in virtualenv :", in_venv)
print("sys.prefix    :", sys.prefix)
print("VIRTUAL_ENV   :", os.environ.get("VIRTUAL_ENV", "(unset)"))
````

## Checking pip and installed packages

`pip list` is available from the shell; from Python you can also import modules and read their `__version__` string.

````python
import importlib, subprocess, sys

for name in ("json", "re", "pathlib"):
    mod = importlib.import_module(name)
    print(f"{name:10s} ok  ({mod.__name__})")

# pip version via subprocess (stdlib only)
r = subprocess.run([sys.executable, "-m", "pip", "--version"],
                   capture_output=True, text=True)
print(r.stdout.strip() or r.stderr.strip())
````

## requirements.txt habit

Pin versions so that teammates reproduce your environment.

````python
from pathlib import Path
requirements = "\n".join([
    "# example only - nothing is installed by this cell",
    "numpy==1.26.4",
    "pandas==2.2.2",
    "matplotlib==3.8.4",
])
Path("requirements_example.txt").write_text(requirements)
print(Path("requirements_example.txt").read_text())
````

## Directory hygiene for a VLSI project

A tidy project layout saves hours later.

````python
from pathlib import Path

root = Path("demo_project")
for sub in ("scripts", "reports", "constraints", "results"):
    (root / sub).mkdir(parents=True, exist_ok=True)

for p in sorted(root.rglob("*")):
    print(p.relative_to(root))
````

## VLSI practice — environment fingerprint

Write a function that prints a short environment fingerprint — the kind you'd attach to a regression run log.

````python
import sys, platform, datetime

def fingerprint():
    return {
        "python": sys.version.split()[0],
        "exe": sys.executable,
        "os": platform.platform(),
        "when": datetime.datetime.now().isoformat(timespec="seconds"),
    }

for k, v in fingerprint().items():
    print(f"{k:7s}: {v}")
````
