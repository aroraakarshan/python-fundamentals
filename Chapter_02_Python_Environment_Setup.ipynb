## Is Python already here?

On most Linux machines — including every VLSI workstation the book targets — Python 3 is pre-installed.

````python
import sys
print('Executable:', sys.executable)
print('Version:   ', sys.version.split()[0])
````

The book recommends **Python 3.11.4 or newer** for VLSI work. This notebook requires **3.9+** as a minimum. If you see anything older, ask IT for the latest Python package path — don't upgrade the system install yourself.

## `pip` — Python's package manager

````python
import subprocess, sys
print(subprocess.run([sys.executable, '-m', 'pip', '--version'],
                     capture_output=True, text=True).stdout.strip())
````

If this prints a version, you're ready. If it says `No module named pip`, run `python -m ensurepip --upgrade` once.

## Picking an editor

| Editor | Best for | Install |
|---|---|---|
| **Jupyter Notebook** | Exploring, plotting (you're in one now) | `pip install notebook` |
| **VS Code** | Day-to-day scripting, remote editing | Download from code.visualstudio.com |
| **PyCharm** | Larger projects, refactoring | Community edition from jetbrains.com |

## Installing a package — the safe pattern

````python
# Shape of a real install (commented so this cell doesn't actually install anything):
# subprocess.run([sys.executable, '-m', 'pip', 'install', '--user', 'requests'])
print('Safe pattern: python -m pip install --user <package>')
````

Rules from the book:

- **Never** `sudo pip install` on a shared VLSI server.
- Prefer `--user` or a virtual environment.
- Record exact versions with `pip freeze > requirements.txt` for production scripts.
