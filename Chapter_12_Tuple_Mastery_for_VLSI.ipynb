## Creating tuples

````python
pin_location = (120, 340)
cell_data = ('INV_X1', 1.2, 0.05)
bbox = (10, 20, 110, 150)
print(pin_location, cell_data, bbox)
````

## Single-element tuple — trailing comma matters

````python
x = (42)
y = (42,)
print(type(x).__name__, type(y).__name__)
````

## Access and unpack

````python
coord = (100, 200)
print(coord[0], coord[1])
name, area, power = ('NAND2_X1', 2.3, 0.08)
print(name, area, power)
````

## Immutability is a feature

Tuples can be dictionary keys; lists cannot. Great for composite keys like timing arcs.

````python
timing_arc = {
    ('A', 'Z'): 0.12,
    ('B', 'Z'): 0.15,
    ('CLK', 'Q'): 0.20,
}
print(timing_arc[('CLK', 'Q')])
for (src, dst), delay in timing_arc.items():
    print(f'{src} → {dst}: {delay} ns')
````

## Lists of tuples — records

````python
pins = [(100, 200), (150, 250), (180, 300)]
for x, y in pins:
    print(f'({x}, {y})')
````

## What you can't do

Attempting to change a tuple raises `TypeError` — which prevents silent bugs.

````python
coord = (100, 200)
try:
    coord[0] = 120
except TypeError as e:
    print('expected error:', e)
````

## VLSI in practice — compute total bbox area

````python
bboxes = [
    (10, 20, 110, 150),
    (200, 100, 260, 220),
    (300, 300, 400, 400),
]
total = 0
for x1, y1, x2, y2 in bboxes:
    area = (x2 - x1) * (y2 - y1)
    total += area
    print(f'bbox ({x1},{y1})-({x2},{y2}) area={area}')
print('total area:', total)
````

## Recap

- Tuples are ordered, immutable — great for fixed records.
- A single-element tuple needs a trailing comma: `(42,)`.
- Tuples can be dictionary keys; lists cannot.
- Unpacking gives readable names: `x, y = coord`.
