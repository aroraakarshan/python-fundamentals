## Why functions?

Functions let you name a block of work once and call it many times. Less duplication, fewer bugs, easier to test.

````python
def greet():
    print('Hello, VLSI engineer!')

greet()
````

## Parameters and `return`

````python
def check_slack(slack):
    if slack < 0:
        return 'Violation'
    elif slack == 0:
        return 'Critical'
    else:
        return 'Safe'

print(check_slack(-0.05))
print(check_slack(0.0))
print(check_slack(0.12))
````

## Default arguments

````python
def log_violation(net, severity='low'):
    print(f'{net} has a {severity} violation')

log_violation('clk')
log_violation('rst', 'high')
````

## Positional vs keyword arguments

Positional arguments match by order. Keyword arguments match by name — clearer for calls with many parameters.

````python
def report_violation(cell, severity):
    print(f'{cell}: {severity}')

report_violation('U1', 'high')
report_violation(severity='medium', cell='U2')
report_violation('U3', severity='low')
````

## Keyword-only arguments with `*`

Anything after `*` must be passed by name — useful to prevent ambiguous calls.

````python
def config(*, mode, threshold):
    print(mode, threshold)

config(mode='setup', threshold=0.15)
try:
    config('setup', 0.15)
except TypeError as e:
    print('expected error:', e)
````

## `*args` — variable positional arguments

````python
def report_cells(*cells):
    for cell in cells:
        print('Checking', cell)

report_cells('U1', 'U2', 'U3')
````

## `**kwargs` — variable keyword arguments

````python
def configure_tool(**settings):
    for key, value in settings.items():
        print(f'{key} set to {value}')

configure_tool(clock='clk', threshold=0.2, mode='hold')
````

## VLSI in practice — classify a list of slacks

````python
def classify(slack, setup_margin=0.0):
    if slack < setup_margin:
        return 'VIOLATION'
    if slack < setup_margin + 0.05:
        return 'MARGINAL'
    return 'OK'

paths = [('P1', -0.04), ('P2', 0.02), ('P3', 0.12), ('P4', 0.0)]
for name, s in paths:
    print(f'{name}: slack={s:+.3f}  -> {classify(s)}')
````

## Recap

- `def name(params):` introduces a function; `return` sends a value back.
- Default values make parameters optional.
- Positional args match by order; keyword args match by name.
- `*args` bundles extra positional, `**kwargs` bundles extra keyword arguments.
