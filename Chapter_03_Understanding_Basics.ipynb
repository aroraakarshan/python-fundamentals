## Variables are just labels

Assignment binds a name to a value. No declaration, no type annotation needed (though type hints are welcome).

````python
cell_name = "INV_X1"
pin_count = 4
is_sequential = False
print(cell_name, pin_count, is_sequential)

# A name can be rebound to a different type — Python is dynamic:
cell_name = 42
print("rebound:", cell_name, type(cell_name).__name__)
````

## Core built-in types

Five types cover 90% of code: int, float, str, bool, and None.

````python
samples = [
    0,
    3.14,
    "hello",
    True,
    None,
]
for v in samples:
    print(f"{v!r:10s} -> {type(v).__name__}")
````

## Truthiness

Empty collections, zero values, and `None` are falsy; almost everything else is truthy.

````python
for v in (0, 0.0, "", [], {}, None, "x", 1, [0]):
    print(f"{v!r:5s} -> {bool(v)}")
````

## Type conversion

Use `int()`, `float()`, `str()`, `bool()` to convert. Conversions can raise — protect with `try/except` when inputs are untrusted.

````python
print(int("42"), float("3.14"), str(100), bool("hi"))

try:
    int("3.14")
except ValueError as e:
    print("oops:", e)
````

## Indentation is syntax

Python uses 4-space indentation to delimit blocks. Mixing tabs and spaces causes ugly errors.

````python
slack = -0.05
if slack < 0:
    print("violation")
    print("fix me")
else:
    print("clean")
````

## Comments and docstrings

`#` starts a single-line comment. Triple-quoted strings at the top of a function serve as docstrings.

````python
def describe_slack(slack):
    """Return a one-word label for a timing slack value."""
    if slack < 0:
        return "violation"
    return "met"

print(describe_slack(-0.01))
print(describe_slack.__doc__)
````

## Naming conventions

`snake_case` for variables and functions, `CamelCase` for classes, `UPPER_CASE` for constants.

````python
MAX_CLOCK_GHZ = 5.0
target_clock  = 1.2
class TimingReport: pass

print(MAX_CLOCK_GHZ, target_clock, TimingReport.__name__)
````

## Common newcomer mistakes

These are the ones that fool everyone at least once.

````python
# 1. '=' vs '==':
x = 5
print("equal?", x == 5)

# 2. Integer vs float division:
print(7 / 2, 7 // 2)

# 3. Mutable default arguments (don't do this):
def wrong(items=[]):
    items.append("x")
    return items
print(wrong(), wrong())  # shared state bug!
````

## VLSI practice — a tiny validation helper

Validate a slack string from a report: convert to float, classify it, and refuse malformed input politely.

````python
def classify_slack(raw):
    try:
        slack = float(raw)
    except (TypeError, ValueError):
        return ("invalid", raw)
    if slack < 0:
        return ("violation", slack)
    if slack == 0:
        return ("critical", slack)
    return ("met", slack)

for v in ("-0.045", "0.12", "n/a", ""):
    print(f"{v!r:10s} -> {classify_slack(v)}")
````
