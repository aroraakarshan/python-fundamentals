## What is a variable?

A variable is a labeled box in memory. Python uses `=` — no `set` or `$`.

````python
voltage = 1.2
cell_name = 'INV_X1'
num_pins = 24
is_clock = True
print(voltage, cell_name, num_pins, is_clock)
````

## Data types

Python figures out the type automatically. The four everyday ones:

- `int` — whole numbers: `24`, `-5`
- `float` — decimals: `1.2`, `-0.045`
- `str` — text: `"U1"`
- `bool` — `True` or `False`

````python
for v in (voltage, cell_name, num_pins, is_clock):
    print(v, '→', type(v).__name__)
````

## Comments — notes for humans

Everything after `#` is ignored. Explain *why*, not *what*.

````python
# Calculate power as voltage * current
current = 0.5
power = voltage * current
print(f'power = {power} W')
````

## What is a script?

A script is just a file ending in `.py` that runs top to bottom. The cells in this notebook are conceptually the same — they execute in order. Chapters 4–18 in the book can all be practised in a single `.py` file.

## What is a function?

A reusable block of code. `def` names it; `return` hands back the result.

````python
def calculate_power(v, i):
    return v * i

print('Power:', calculate_power(1.1, 0.3))
````

## What is a module?

A file full of functions and variables. `import` brings it in. Python ships with many.

````python
import math
print(math.sqrt(16))
print(math.pi)
````

## What is a package?

A group of related modules installed as a unit (`pip install ...`). `import` brings it in, same as a module.

````python
import json
cell_info = {'name': 'INV_X1', 'area': 1.2, 'power': 0.05}
print(json.dumps(cell_info, indent=2))
````
