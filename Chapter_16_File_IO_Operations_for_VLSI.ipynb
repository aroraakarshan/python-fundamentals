## Open a file — always use `with`

`with` guarantees the file closes, even on error.

````python
from pathlib import Path
Path('file.txt').write_text('hello VLSI engineer\n')

with open('file.txt', 'r') as f:
    content = f.read()
print(content)
````

## File modes

`'r'` read, `'w'` write (overwrite), `'a'` append, `'x'` create-new, `'b'` binary, `'t'` text (default).

## Line-by-line — the memory-friendly pattern

````python
from pathlib import Path
Path('sta.log').write_text(
    'INFO starting run\n'
    'slack -0.03 setup VIOLATION\n'
    'slack 0.12 hold met\n'
    'INFO done\n'
)
with open('sta.log') as f:
    for line in f:
        print(line.rstrip())
````

## Read all lines into a list

````python
with open('sta.log') as f:
    lines = f.readlines()
print(len(lines), 'lines')
print(lines[:2])
````

## VLSI use case — filter a report

````python
with open('sta.log') as f:
    for line in f:
        if 'VIOLATION' in line:
            print('Hold/Setup:', line.strip())
````

## Writing files

````python
with open('summary.txt', 'w') as f:
    f.write('Slack Violations: 3\n')

with open('summary.txt', 'a') as f:
    f.write('New check complete.\n')

print(Path('summary.txt').read_text())
````

## VLSI use case — write a list of nets

````python
nets = ['clk', 'reset', 'vdd']
with open('critical_nets.txt', 'w') as f:
    for net in nets:
        f.write(net + '\n')

print(Path('critical_nets.txt').read_text())
````

## Compressed files with `gzip`

STA dumps are often `.gz`. Use `'rt'` for text mode.

````python
import gzip
from pathlib import Path
with gzip.open('sta.rpt.gz', 'wt') as f:
    f.write('slack -0.04 VIOLATION\n')
    f.write('slack 0.02 OK\n')

with gzip.open('sta.rpt.gz', 'rt') as f:
    for line in f:
        if 'VIOLATION' in line:
            print(line.strip())
````

## `pathlib` — the modern path API

Object-oriented, chainable, cross-platform.

````python
from pathlib import Path
p = Path('critical_nets.txt')
print('exists:', p.exists())
print('is file:', p.is_file())
print('size:', p.stat().st_size, 'bytes')
print('name:', p.name, 'suffix:', p.suffix)

logs = Path('logs_demo')
logs.mkdir(exist_ok=True)
(logs / 'a.rpt').write_text('one\n')
(logs / 'b.rpt').write_text('two\n')
for f in sorted(logs.glob('*.rpt')):
    print('found:', f.name)
````

## VLSI use case — check for required input files

````python
from pathlib import Path
Path('netlist.v').write_text('// stub\n')
Path('cells.lib').write_text('// stub\n')
required = ['netlist.v', 'constraints.sdc', 'cells.lib']
missing = [f for f in required if not Path(f).exists()]
print('missing:', missing)
````

## Recap

- Always open files with `with` — close is automatic.
- `'r'/'w'/'a'` are the everyday modes; add `'t'/'b'` for text/binary.
- Iterate line-by-line for large files; `readlines()` loads everything.
- `pathlib` is the clean, object-oriented way to work with paths.
- Use `gzip.open(..., 'rt')` for `.gz` text files.
