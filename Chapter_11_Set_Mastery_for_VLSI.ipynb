## Creating sets

Use `{...}` or `set(iterable)`. Remember: `{}` is an empty dict, not set!

````python
cell_types = {"INV_X1", "NAND2_X1", "DFF_X1"}
seen_layers = set(["M1", "M2", "M1", "M3"])
print(cell_types)
print("deduped :", seen_layers)
print("empty   :", set())
````

## Adding and removing

`add` inserts, `discard` removes silently, `remove` raises `KeyError`.

````python
layers = {"M1", "M2", "M3"}
layers.add("M4")
layers.discard("M9")
try:
    layers.remove("M9")
except KeyError as e:
    print("remove raised:", e)
print(layers)
````

## Fast membership

`in` on a set is O(1) on average — way faster than on a list.

````python
used = {"M1", "M2", "M3", "M5", "M6"}
for L in ("M1", "M4", "M6", "M9"):
    print(f"{L}: {'used' if L in used else 'unused'}")
````

## Set algebra

`|` union, `&` intersection, `-` difference, `^` symmetric difference.

````python
layout  = {"M1", "M2", "M3", "M4"}
allowed = {"M1", "M2", "M4", "M5"}
print("union     :", layout | allowed)
print("common    :", layout & allowed)
print("illegal   :", layout - allowed)
print("mismatch  :", layout ^ allowed)
````

## Deduplicating lists

One-liner: `sorted(set(iterable))`. If you must preserve original order, track a `seen` set.

````python
nets = ["clk", "data", "clk", "reset", "data", "clk"]
print(sorted(set(nets)))

seen, ordered = set(), []
for n in nets:
    if n not in seen:
        ordered.append(n); seen.add(n)
print(ordered)
````

## Frozen sets

`frozenset` is immutable, so it can be a dict key or live in another set.

````python
power_domains = {
    frozenset({"VDD_CORE","VSS"}): "logic",
    frozenset({"VDD_IO",  "VSS"}): "io",
    frozenset({"VDD_MEM", "VSS"}): "memory",
}
for domain, kind in power_domains.items():
    print(sorted(domain), "->", kind)
````

## VLSI practice — DRC layer audit

Compare used vs allowed layers and report illegal, missing, and Jaccard similarity.

````python
used    = {"M1","M2","M3","M5","VIA1","VIA2"}
allowed = {"M1","M2","M3","M4","VIA1","VIA2","VIA3"}

illegal = used - allowed
missing = allowed - used
common  = used & allowed
jacc    = len(common) / len(used | allowed)

print(f"illegal : {sorted(illegal)}")
print(f"missing : {sorted(missing)}")
print(f"common  : {sorted(common)}")
print(f"jaccard : {jacc:.2f}")
````

## Recap

- Sets store unique, hashable items with O(1) membership.
- `|`, `&`, `-`, `^` give you union/intersection/difference/xor.
- `sorted(set(x))` is the cleanest dedup one-liner.
- `frozenset` is immutable and hashable — use it as dict keys.
