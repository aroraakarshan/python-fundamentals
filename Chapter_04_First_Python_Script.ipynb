## The book's first script

The classic three lines: define voltage and current, multiply, print. As a file you'd save this as `power_calc.py` and run `python3 power_calc.py`.

````python
# This script calculates power from voltage and current

# Define input values
voltage = 1.2   # in volts
current = 0.5   # in amperes

# Calculate power
power = voltage * current

# Print the result
print('Voltage:', voltage, 'V')
print('Current:', current, 'A')
print('Power:', power, 'W')
````

## Reading it line by line

- Lines starting with `#` are comments.
- `voltage` and `current` are **variables** holding floats.
- `power = voltage * current` is the calculation.
- `print()` sends output to the terminal.

## The same script in Tcl and Perl

Reference only — don't run. Feel how much ceremony each extra character costs.

**Tcl:**
```tcl
set voltage 1.2
set current 0.5
set power [expr {$voltage * $current}]
puts "Power: $power W"
```

**Perl:**
```perl
$voltage = 1.2;
$current = 0.5;
$power = $voltage * $current;
print "Power: $power W\n";
```

Python wins on clarity: no `expr`, no `$`, no trailing semicolons.

## Common mistakes to avoid

- **Forgetting `.py` in the filename.** `python3` won't find your script.
- **Wrong Python on PATH.** Run `python3 --version` to check.
- **Indentation inside blocks.** Python uses whitespace to define structure.

````python
# Formatted output with an f-string — more on these in Chapter 7
voltage, current = 1.2, 0.5
power = voltage * current
print(f'V={voltage} V, I={current} A → P={power:.3f} W')
````

## VLSI in practice — a tiny rail power table

````python
budget_W = 0.75
rails = [
    ('VDD_CORE', 0.8, 0.5),
    ('VDD_IO',   1.8, 0.3),
    ('VDD_SRAM', 0.9, 1.2),
]
print(f"{'rail':10s} {'V':>4s} {'I':>4s} {'P(W)':>6s}  status")
for rail, v, i in rails:
    p = v * i
    status = 'over_budget' if p > budget_W else 'ok'
    print(f'{rail:10s} {v:4.2f} {i:4.2f} {p:6.3f}  {status}')
````
