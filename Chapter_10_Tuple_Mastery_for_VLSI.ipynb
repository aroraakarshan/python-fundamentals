## Creating tuples

Parentheses are optional; the comma makes a tuple.

````python
row = ("reg_a2b", -0.045, "ns")
single = (42,)          # trailing comma matters!
empty = ()
print(row, single, empty, type(single))
````

## Immutability

Tuples can't be changed, which makes them safer to share between functions and usable as dict keys.

````python
t = (1, 2, 3)
try:
    t[0] = 99
except TypeError as e:
    print("error:", e)
````

## Indexing and slicing

Same syntax as lists. `len`, `in`, `count`, `index` also work.

````python
path = ("top", "cpu", "alu", "adder_0")
print(path[0], path[-1])
print("has cpu:", "cpu" in path)
print(path[:2])
````

## Unpacking

Assign multiple variables at once. `*rest` captures the middle or tail.

````python
row = ("reg_a2b", -0.045, "ns")
name, slack, unit = row
print(name, slack, unit)

first, *middle, last = (1, 2, 3, 4, 5)
print(first, middle, last)
````

## Swap without a temp

Tuple packing/unpacking makes swap a one-liner.

````python
a, b = 1, 2
a, b = b, a
print(a, b)
````

## Tuples as dict keys

Because tuples are immutable (if their contents are), they can key a dict.

````python
timing = {
    ("reg_a", "reg_b"): -0.045,
    ("reg_b", "reg_c"):  0.120,
    ("reg_a", "reg_c"): -0.020,
}
print(timing[("reg_a", "reg_b")])
for edge, slack in timing.items():
    src, dst = edge
    print(f"{src} -> {dst}  slack={slack:+.3f}")
````

## NamedTuple for readability

When you find yourself writing `row[2]`, reach for `NamedTuple` so you can write `row.unit` instead.

````python
from typing import NamedTuple

class Path(NamedTuple):
    start: str
    end:   str
    slack: float
    unit:  str = "ns"

p = Path("reg_a", "reg_b", -0.045)
print(p)
print(p.start, p.slack, p.unit)
````

## VLSI practice — tuple-keyed timing arc table

Store per-corner timing arcs keyed by `(cell, from_pin, to_pin, corner)` and query.

````python
arcs = {
    ("INV_X1", "A", "Y", "ff_0p88v_125c"): 0.045,
    ("INV_X1", "A", "Y", "ss_0p72v_m40c"): 0.120,
    ("NAND2_X1", "A", "Y", "ff_0p88v_125c"): 0.055,
}

def arc(cell, fp, tp, corner, default=None):
    return arcs.get((cell, fp, tp, corner), default)

print("INV_X1 A->Y fast :", arc("INV_X1", "A", "Y", "ff_0p88v_125c"))
print("INV_X1 A->Y slow :", arc("INV_X1", "A", "Y", "ss_0p72v_m40c"))
print("DFF  Q->D slow   :", arc("DFF_X1", "Q", "D", "ss_0p72v_m40c", "?"))
````

## Recap

- Tuples are immutable, ordered, and lightweight records.
- Unpack with `a, b, c = tup` — use `*rest` for variable-length tails.
- Because they're hashable, tuples make excellent compound dict keys.
- `NamedTuple` swaps positional access for self-documenting field names.
